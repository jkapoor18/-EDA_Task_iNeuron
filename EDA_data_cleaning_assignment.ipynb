# now lets clean the dataset or replacing unstructured values

```python
data_copy=data_copy.rename(columns={"39":"Age"})
data_copy
```

# Here updating the column name

```python
data_copy=data_copy.rename(columns={' State-gov':'Workclass'})
data_copy=data_copy.rename(columns={' 40':'Hour'})
data_copy=data_copy.rename(columns={' Never-married':'Marital-status'})
data_copy=data_copy.rename(columns={' Adm-clerical':'Occupation'})
data_copy=data_copy.rename(columns={' Not_in_family':'Relationship'})
data_copy=data_copy.rename(columns={' White':'Race'})
data_copy=data_copy.rename(columns={' Male':'Gender'})
data_copy=data_copy.rename(columns={' United-States':'Country'})
data_copy=data_copy.rename(columns={' <=50K':'Salary'})
data_copy=data_copy.rename(columns={' 2174':'Capital-gain'})
data_copy=data_copy.rename(columns={' 0':'Capital-loss'})
data_copy=data_copy.rename(columns={' 77516':'Fcode'})
data_copy=data_copy.rename(columns={' 13':'Education-num'})
data_copy=data_copy.rename(columns={' Bachelors':'Education'})

data_copy.head()
```

```python
data_copy
```

```python
data_copy['Salary'].unique()
```

```python
char_to_remove=[">","<","="]
cols_to_clean=['Salary']
for item in char_to_remove:
    for col in cols_to_clean:
        data_copy[col]=data_copy[col].str.replace(item,"")
```

```python
data_copy.head()
```

```python
char_to_remove=["K"]
cols_to_clean=['Salary']
for item in char_to_remove:
    for col in cols_to_clean:
        data_copy[col]=data_copy[col].str.replace(item,"000")
```

# now ploting graph

```python
plt.figure(figsize=(12,6))
sns.histplot(data_copy['Age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

# it shows that the majority of the individuals in the dataset are between the ages of 20 and 50.

```python
# Distribution of workclass
plt.figure(figsize=(12,6))
sns.countplot(x='Workclass', data=data_copy)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Distribution of Workclass')
plt.xticks(rotation=45)
plt.show()
```

# majority of workclass in private sector

```python
# Distribution of race
plt.figure(figsize=(12,6))
sns.countplot(x='Race', data=data_copy)
plt.xlabel('Race')
plt.ylabel('Count')
plt.title('Distribution of Race')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of sex
plt.figure(figsize=(12,6))
sns.countplot(x='Gender', data=data_copy)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of native country
plt.figure(figsize=(12,6))
sns.countplot(y='Country', data=data_copy)
plt.xlabel('Count')
plt.ylabel('Country')
plt.title('Distribution of Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between education and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Education', hue='Salary', data=data_copy)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Relationship between Education and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

```python
# Relationship between workclass and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Workclass', hue='Salary', data=data_copy)
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.title('Relationship between Workclass and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```

```python
# Relationship between marital status and salary
plt.figure(figsize=(12,6))
sns.countplot(x='Marital-status', hue='Salary', data=data_copy)
plt.xlabel('Marital-status')
plt.ylabel('Count')
plt.title('Relationship between Marital-status and Salary')
plt.xticks(rotation=45)
plt.legend(title='Salary', loc='upper right')
plt.show()
```
